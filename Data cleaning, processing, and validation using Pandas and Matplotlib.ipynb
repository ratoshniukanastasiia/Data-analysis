```python
import pandas as pd
```

```python
# Uploading the file
apl = pd.read_csv('applications.csv')
apl
```

```python
# Uploading another file
ind = pd.read_csv('industries.csv')
ind.head(10)
```

Removing duplicates in applications DF

```python
# removing duplicates in applications DF
apl_uniq = apl.drop_duplicates(subset=['applicant_id'])
apl_uniq
```

The replacement of NaN values to 0 in External Rating of DF and to "Середня" in Education level

```python
# The replacement of NaN values to 0 in External Rating of DF and to "Середня" in Education level
apl_uniqu = apl_uniq.fillna({'External Rating':0, 'Education level':'Середня'})
apl_uniqu.head()
```

Joinning two DFs using function 'merge'

```python
# Joinning two DFs
merged_df = pd.merge(apl_uniqu, ind, on = 'Industry', how = 'outer')
merged_df.head()
```

```python
# Info about df
merged_df.info()
```

Finding the Score range in DF

```python
# Finding the Score range in DF
print(merged_df['Score'].min())
print(merged_df['Score'].max())
```

Creating the Rating table using a cycle and conditions. Outputing of the changed dataframe.

```python
# Creating the Rating table

raiting = []

for index, row in merged_df.iterrows():
    rating = 0
    
    if pd.isna(row['Amount']) or row['External Rating'] == 0:
        rating = 0 # Raiting is 0 score
    else:
        if 35 <= row['Age'] <= 55:
            rating += 20 # if age betwen 35 and 55 than add 20 scores
            
        applied_date = pd.to_datetime(row['Applied at'])
        if applied_date.dayofweek != 6 and applied_date.dayofweek != 7:
            rating += 20 # if applied date is a weekday than add scores
            
        if 'Married' in row['Marital status']:
            rating += 20 # if 'Married' than add 20 scores
    
        if row['Location'] == 'Київ чи область':
            rating += 10 # if Location is 'Київ чи область' than add 10 scores
    
        if not pd.isna(row['Score']):
            rating += row['Score'] # add Score values to raiting
    
        if row['External Rating'] >= 7:
            rating += 20 # if External Rating more or equal to 7 than add 20 scores
        
        elif row['External Rating'] <= 2:
            rating -= 20 # if External Rating less or equal to 2 than minus 20 scores
    
    # add elements to a list using the append method.
    raiting.append(rating)

# Add the rating column to DataFrame
merged_df['Rating'] = raiting
merged_df['Rating'] = merged_df['Rating'].clip(upper=100) # clipping rating up to 100 scores
merged_df.head()
```

Sotring our DF by the Rating over 0

```python
# Showing values with the next condition: Rating > 0:
positive_rait = merged_df[merged_df['Rating'] > 0]
positive_rait.head()
```

In case if you want to have a look at the top Rating respondents, I sorted the DF as follows.

```python
# Top Rating
positive_rait.sort_values(by='Rating', ascending=False).head()
```

DF which grouped by an applied week and building the chart 

```python
# import matplotlib 
import matplotlib.pyplot as plt

def addlabels(x, y):
    for i in range(len(x)):
        # Round the value to 1 decimal places before displaying it
        rounded_value = round(y[i], 1)
        plt.text(y[i] // 2, i, str(rounded_value), ha='center', va='center') # the functing to show values inside the bar chart

# groupping values by numbers of week and getting the average values
positive_rait1=positive_rait.copy()
positive_rait1['Applied at'] = pd.to_datetime(positive_rait1['Applied at'])
avg_rait_w = positive_rait1.groupby(positive_rait1['Applied at'].dt.strftime('%Y-%W'))['Rating'].mean() # week starts from Monday

# building the chart
plt.style.use('tableau-colorblind10')

plt.figure(figsize=(8, 6))

plt.barh(avg_rait_w.index, avg_rait_w.values, color='cornflowerblue', edgecolor='navy', linewidth=1,) #building the bar chart

addlabels(avg_rait_w.index, avg_rait_w.values)

plt.xlabel('Week number')
plt.ylabel('Mean Rating')
plt.title('Average Rating by Week')
plt.show()
```
